## 第一周/第三题-matplotlib/自定义图表元素

```python
# 使用matplotlib库绘制函数 y=sin(x) 的图像，并同时满足以下所有要求：
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
rcParams['font.sans-serif']=['SimHei']
rcParams['axes.unicode_minus']=False
# - 曲线线型为虚线；
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,color='red',label='sin(x)',linewidth=2,linestyle='--')
# - y 轴范围固定为 ([-1.5,,1.5])；
plt.ylim(-1.5,1.5)
# - x 轴范围为 ([0,10])；
plt.xlim(0,10)
# - 坐标轴标签分别为：x 轴 “x”，y 轴 “y”；
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
# - 图表标题为 “三角函数”（注意解决中文乱码问题）；
plt.title('三角函数',color='purple',fontsize=15)
# - 显示网格线；
plt.grid(True,linestyle='-',color='black',alpha=0.3)
# - 给曲线添加图例（标签为 `sin(x)`），并显示在图中合适位置。
plt.legend(loc='upper right',fontsize=10)
# **输出：** 一张满足以上全部要求的 Matplotlib 图像。
plt.show()
```
