## 第一周/第四题-matplotlib-numpy/绘制气泡散点图

```python
# 请使用NumPy和Matplotlib生成随机数据并绘制一幅“气泡散点图”
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
rcParams['axes.unicode_minus']=False
# 1. 随机生成 100 个点的坐标：
np.random.seed(10)
#    - `x` 与 `y` 分别来自标准正态分布。
x=np.random.randn(100)
y=np.random.randn(100)

plt.figure(figsize=(5,5))
# 2. 随机生成与每个点对应的两个属性：
#    - `colors`：长度为 100，取值在 `[0, 1)`，符合标准正态分布，用于映射颜色。
colors=np.random.rand(100)
#    - `sizes`：长度为 100，取值在 `[0, 1000)`，符合标准正态分布，用于设置点面积。
sizes=np.random.rand(100)*1000
# 3. 使用 `plt.scatter` 绘图，参数要求：
#    - `c=colors`，`s=sizes`，`alpha=0.3`，`cmap='viridis'`。
plt.scatter(x,y,c=colors,cmap='viridis',s=sizes,alpha=0.3)
# 4. 添加颜色条（色阶说明）。
plt.colorbar()
# **输出**:运行后应显示一幅带有颜色条的气泡散点图。
plt.tight_layout()
plt.show()
```
