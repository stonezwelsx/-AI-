### 第一周/第二题-pandas-numpy/列和最小列名

```python
# 使用pandas、NumPy请完成如下任务：
import pandas as pd
import numpy as np
# 1. 生成一个 5×5 的随机浮点数数组，元素取自区间 ([0,1))
np.random.seed(0)
arr=np.random.uniform(0,1,(5,5))
# 2. 将其转换为一个 `pandas.DataFrame`，列名依次为 `a, b, c, d, e`。
arr_pd=pd.DataFrame(arr,columns=['a','b','c','d','e'])
# 3. 将该表完整打印到控制台。
print(arr_pd)
# 4. 计算每一列的列和，并输出“列和最小”的那一列的列名（字符串形式）。
r=arr_pd[['a','b','c','d','e']].sum()
r=''.join(map(str,arr_pd.columns[r==r.min()].tolist()))
r
```
